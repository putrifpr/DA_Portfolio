**13. How many people have a credit card that expires in 2020?**

```python
def exp_cc():
  count = 0
  for date in sample['CC Exp Date']:
    if date.split('/')[1] == '20':
      count = count + 1
  print(count)

exp_cc()
```

```python
sample[sample['CC Exp Date'].apply(lambda x:x[3:] == '20')]
```

```python
len(sample[sample['CC Exp Date'].apply(lambda x:x[3:] == '20')])
```

**14. Top 5 most popular email providers (e.g. gmail.com, yahoo.com, etc)**

```python
# Cara 1

list_1 = []
for email in sample['Email']:
  list_1.append(email.split('@')[1])

sample['Email providers'] = list_1
```
